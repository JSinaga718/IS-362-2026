```python
windiest_day = weather[
    (weather['year'] == 2013) &
    (weather['month'] == 2) &
    (weather['day'] == 12)
]

avg_wind = windiest_day.groupby('origin')['wind_speed'].mean().sort_values(ascending=False)
print(avg_wind)
```
