```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cumulative_miles = pd.Series([
    55,    # Day 1
    120,   # Day 2
    185,   # Day 3
    260,   # Day 4
    330,   # Day 5
    405,   # Day 6
    475,   # Day 7
    545,   # Day 8
    620,   # Day 9
    695,   # Day 10
    765,   # Day 11
    840,   # Day 12
    910,   # Day 13
    985,   # Day 14
    1060,  # Day 15
    1130,  # Day 16
    1205,  # Day 17
    1280,  # Day 18
    1350,  # Day 19
    1420,  # Day 20
    1500   # Day 21 
], index=range(1, 22), name='Cumulative Miles')

cumulative_miles.index.name = 'Day'

print("Cumulative Odometer Readings:")
print(cumulative_miles.to_string())
print(f"\nTotal number of days: {len(cumulative_miles)}")
print(f"Final odometer reading: {cumulative_miles.iloc[-1]} miles")

daily_miles = cumulative_miles.diff()

daily_miles.iloc[0] = cumulative_miles.iloc[0]
daily_miles.name = 'Daily Miles'

print("\nDaily Miles Ridden:")
print(daily_miles.to_string())

trip_summary = pd.DataFrame({
    'Daily Miles': daily_miles,
    'Cumulative Miles': cumulative_miles
})

print("\nComplete Trip Summary:")
print(trip_summary)
print(f"\nTotal Distance: {cumulative_miles.iloc[-1]} miles")
print(f"Longest Day: {daily_miles.max():.0f} miles (Day {daily_miles.idxmax()})")
print(f"Shortest Day: {daily_miles.min():.0f} miles (Day {daily_miles.idxmin()})")
```
